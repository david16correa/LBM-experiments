```python
nParticles = len(particleDf.particleId.unique())
particleTime = particleDf.time.unique()
```

```python
fig, axes = plt.subplots(1,2,figsize=(6,2.7))

for Id in np.arange(nParticles):
    axes[0].plot(particleTime, particleDf.query(f"particleId == {Id+1}").position_x.values, )
axes[0].set_title("position -- $x$")
    
for Id in np.arange(nParticles):
    axes[1].plot(particleTime, np.sqrt(
            particleDf.query(f"particleId == {Id+1}").velocity_x.values**2 + 
            particleDf.query(f"particleId == {Id+1}").velocity_y.values**2
        )
    )
axes[1].set_title("speed -- $|\\mathbf{u}|$")
axes[1].set_ylim(0,1e-3)
axes[1].yaxis.tick_right()  # Moves y-axis ticks to the right
axes[1].set_yticks([0, 0.5e-3, 1e-3])
axes[1].set_yticklabels(["0", "", "$10^{-3}$"])
    
plt.show()
```

## Comparison

```python
fig, axes = plt.subplots(1,2,
    figsize=(9,2.7),
    gridspec_kw={'width_ratios': [3, 1]}  # Left plot wider than right plot
)

src_ns = ['06','07','08','09','10','11'] + [f'{Id}' for Id in range(13,int(src_n)+1)] # src 12 was a sanity check and shall not be used here
trainLen = 1
avgSpeeds = []

cmap = plt.get_cmap('Spectral')
norm = plt.Normalize(1, len(src_ns))

for n in src_ns:
    auxDir = f'../data.lbm/{n}/'
    auxDf = pd.read_csv(auxDir+"particlesTrj.csv")

    speed = np.sqrt(
        auxDf.query(f"particleId == {trainLen}").velocity_x.values**2 + 
        auxDf.query(f"particleId == {trainLen}").velocity_y.values**2
    )[:len(particleTime)]
    
    axes[0].plot(particleTime, speed, color = cmap(norm(trainLen)))

    avgSpeeds.append(speed[round(len(speed)/2):].mean())
    trainLen+=1
    
axes[0].set_title("speed of squirmer")
axes[0].set_xlabel("$t ~ (\\mu \\mathrm{s})$")
axes[0].set_ylim(0,1e-3)
axes[0].set_ylabel("$v ~ (\\mu \\mathrm{m} / \\mu \\mathrm{s})$")
axes[0].set_yticks([0, 0.5e-3, 1e-3])
axes[0].set_yticklabels(["0", "", "$10^{-3}$"])

axes[1].scatter(range(1,len(src_ns)+1), avgSpeeds, c=range(1,len(src_ns)+1), cmap=cmap)  # or 'viridis', 'plasma', whatever you like
axes[1].set_xlim(0,len(src_ns)+1)
axes[1].set_xticks(range(1,len(src_ns)+1))
axes[1].set_xticklabels(['0']+['' for _ in range(2,len(src_ns))]+[str(len(src_ns)-1)])
axes[1].set_xlabel('$n$-beads')
axes[1].set_ylim(0,1e-3)
axes[1].set_yticks([0, 0.5e-3, 1e-3])
# axes[1].tick_params(labelbottom=False)
axes[1].tick_params(labelleft=False)

# cbar_ax = fig.add_axes([.925, 0.15, 0.01, 0.7])  # [left, bottom, width, height] for the colorbar
# sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
# sm.set_array([])
# cbar = fig.colorbar(sm, cax=cbar_ax, alpha=0.85)
# cbar.set_label(label='train length', fontsize=16)
# # cbar.ax.tick_params(labelsize=15)

# custom_ticks = np.array([1, len(src_ns)])
# cbar.set_ticks(custom_ticks)

plt.subplots_adjust(wspace=0.05)  # smaller wspace = less horizontal space between plots

plt.show()
```

# stability analysis

```python
def avgSpeed(df):
    return np.sqrt(df.fluidVelocity_x**2 + df.fluidVelocity_y**2).mean()
```

```python
try:
    avgFluidSpeed 
except NameError:
    avgFluidSpeed = []
    
try:
    time 
except NameError:
    time = []

for tick in ticks[len(time)::]:
    auxDf = pd.read_csv(dataDir + f"fluidTrj_{tick}.csv")
    avgFluidSpeed.append(avgSpeed(auxDf))
    time.append(auxDf.time.values[0])
    print(f'\r{tick}', end='', flush=True)
print('\r'+' '*10, end='', flush=True)
```

```python
fig, ax = plt.subplots(figsize=(3,2.7))

ax.plot(time, avgFluidSpeed)
ax.set_xlabel("$t ~ (\\mu \\mathrm{s})$")
ax.set_ylabel("$|u| ~ (\\mu \\mathrm{m} / \\mu \\mathrm{s})$")

plt.show()
```
