# particles

The WCA interaction tries to keep the distance between the squirmers greater than or equal to twice the sum of their radii, plus the lattice paramter:
$$
    r_{12} \geq 2R + \Delta x
$$

> **Note:** $\mathbf{r}_{ij} \equiv \mathbf{x}_i - \mathbf{x}_j$ is the separation vector between particles $i$ and $j$.

```python
x1 = [
    particleDf.query("particleId == 1").position_x.values,
    particleDf.query("particleId == 1").position_y.values,
]
x2 = [
    particleDf.query("particleId == 2").position_x.values,
    particleDf.query("particleId == 2").position_y.values,
]

r12 = [x1[index] - x2[index] for index in np.arange(len(x1))]

particleTime = particleDf.time.unique()

latticeParameter = fluidDf.coordinate_x.unique()[1] - fluidDf.coordinate_x.unique()[0]
```

```python
fig, axes = plt.subplots(figsize=(3,2.7))

axes.plot(
    particleTime,
    np.sqrt(r12[0]**2 + r12[1]**2)
)
# axes[0].set_ylim(-2-latticeParameter, -2+latticeParameter)
minId = np.argmin(np.sqrt(r12[0]**2 + r12[1]**2))
axes.axvline(particleTime[minId], color='gray', linestyle='--', alpha = 0.5, zorder=-1)
axes.axhline(4, color='r', linestyle='--', alpha = 0.5, zorder=-1, label='$2R$')
axes.axhline(4+latticeParameter, color='purple', linestyle='--', alpha = 0.5, zorder=-1, label='$2R + \\Delta x$')
axes.set_xlabel("$t ~ (\\mu \\mathrm{s})$")
axes.set_title("${r}_{12} ~ (\\mu \\mathrm{m})$")

axes.legend()

plt.show()
```

# Stability analysis

```python
def avgSpeed(df):
    return np.sqrt(df.fluidVelocity_x**2 + df.fluidVelocity_y**2).mean()
```

```python
try:
    avgFluidSpeed 
except NameError:
    avgFluidSpeed = []
    
try:
    time 
except NameError:
    time = []

for tick in ticks[len(time)::]:
    auxDf = pd.read_csv(dataDir + f"fluidTrj_{tick}.csv")
    avgFluidSpeed.append(avgSpeed(auxDf))
    time.append(auxDf.time.values[0])
    print(f'\r{tick}', end='', flush=True)
print('\r'+' '*10, end='', flush=True)
```

```python
fig, ax = plt.subplots(figsize=(3,2.7))

ax.plot(time, avgFluidSpeed)
ax.set_xlabel("$t ~ (\\mu \\mathrm{s})$")
ax.set_ylabel("$|u| ~ (\\mu \\mathrm{m} / \\mu \\mathrm{s})$")

plt.show()
```
