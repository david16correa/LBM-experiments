# new Najafi-Golestanian

I'm now using my new molecular dynamics powers to reimplement the Najafi-Golestanian microswimmer. It consists of a polar bond meant to ensure the swimmer is kept straight, and two dynamic linear bonds whose equilibrium distance is time-dependent. This is then an active swimmer, an not a driven one.

This is a shorter simulation; it's meant to help me study the hydrodynamics around the swimmer.

# stability analysis

```python
def avgSpeed(df):
    return np.sqrt(df.fluidVelocity_x**2 + df.fluidVelocity_y**2).mean()
```

```python
try:
    avgFluidSpeed 
except NameError:
    avgFluidSpeed = []
    
try:
    time 
except NameError:
    time = []

for tick in ticks[len(time)::]:
    auxDf = pd.read_csv(dataDir + f"fluidTrj_{tick}.csv")
    avgFluidSpeed.append(avgSpeed(auxDf))
    time.append(auxDf.time.values[0])
    print(f'\r{tick}', end='', flush=True)
print('\r'+' '*10, end='', flush=True)
```

```python
fig, ax = plt.subplots(figsize=(3,2.7))

ax.plot(time, avgFluidSpeed)
ax.set_xlabel("$t ~ (\\mu \\mathrm{s})$")
ax.set_ylabel("$|u| ~ (\\mu \\mathrm{m} / \\mu \\mathrm{s})$")

plt.show()
```
